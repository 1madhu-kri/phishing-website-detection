```python
import pandas as pd
import joblib
import os
import shutil

rf_path = os.path.join("..", "notebook", "saved_models", "random_forest.joblib")
log_reg_path = os.path.join("..", "notebook", "saved_models", "logistic_regression.joblib")
dt_path = os.path.join("..", "notebook", "saved_models", "decision_tree.joblib")

loaded_rf = joblib.load(rf_path)
loaded_log_reg = joblib.load(log_reg_path)
loaded_dt = joblib.load(dt_path)

print("All models imported successfully and ready for production!")
```

```python
source_folder = os.path.join("..", "notebook", "saved_models")
destination_folder = "."  # "." means the current folder where this notebook present

files_to_move = ["decision_tree.joblib", "logistic_regression.joblib", "random_forest.joblib"]

print("--- Moving Files Programmatically ---")

for file_name in files_to_move:
    source_path = os.path.join(source_folder, file_name)
    destination_path = os.path.join(destination_folder, file_name)
    
    if os.path.exists(source_path):
        shutil.move(source_path, destination_path)
        print(f"Moved: {file_name} -> {destination_folder}/")
    else:
        print(f"⚠️ Could not find: {source_path} (It might already be moved!)")

print("\nDone! Check your sidebar now.")
```

```python
csv_path = os.path.join("..", "dataset1.csv")

df = pd.read_csv(csv_path)

X_real = df.drop(columns=['Result']) 

sample_row = X_real.iloc[[0]]

prediction = loaded_rf.predict(sample_row)
probabilities = loaded_rf.predict_proba(sample_row)

prediction = loaded_log_reg.predict(sample_row)
probabilities = loaded_log_reg.predict_proba(sample_row)

prediction = loaded_dt.predict(sample_row)
probabilities = loaded_dt.predict_proba(sample_row)

print("--- Test Prediction Success ---")
print(f"Predicted Class Output: {prediction[0]}")
print(f"Confidence Scores (Phishing vs Legitimate): {probabilities[0]}")
```
